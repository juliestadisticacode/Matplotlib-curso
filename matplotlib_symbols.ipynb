# Markers disponibles
Vamos a tratar los markers disponibles para gráficos en matplotlib

```python
# Importando librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
# en https://matplotlib.org/stable/api/markers_api.html podemos consultar los markers disponibles
# voy a poner algunos más
idx = 0
for marker in ('.', ',', 'o', 'v', '^', '>', '<', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_'):
    plt.plot(idx, 0, ''.join(('r',marker)))
    idx += 1
plt.xlim(-0.5,idx+0.5)
```
