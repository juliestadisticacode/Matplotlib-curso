```python
# Importando librer√≠as
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# Configurando densidad de puntos(DPI) y figure size
plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Como usar los colores adicioneles
import matplotlib.colors as mcolors

#mcolors.BASE_COLORS #these colors can be called with a single character
#mcolors.TABLEAU_COLORS #the default color cycle colors
#mcolors.CSS4_COLORS #named colors also recognized in css
#mcolors.XKCD_COLORS #named colors from the xkcd survey
```

```python
plt.plot([1,2], lw=4, c='xkcd:baby poop green')
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

A = 1.
for color in mcolors.TABLEAU_COLORS:
    plt.plot(x, A * y, c=color, label=color)
    A *= 0.8
plt.legend()
```

```python
'''
Usando tuplas RGB o c√≥digos hex como en html. Las tupla RGB deben de darse en valores entre 0 y 1 por tanto hay que dividir los valores originales entre 255, mientra que el hex code se pasa como string. Podemos obtener los colores deseados buscando color picker en google y obtendremos una paleta de 16 millones de colores'''
plt.plot(x, y, c = (0.84, 0.16, 0))
plt.plot(x, -y, c = '#A50062')
```

```python
# intercambiando formatos
print(mcolors.to_rgb('teal'))
(0.0, 0.5019607843137255, 0.5019607843137255)
print(mcolors.to_hex('chartreuse'))
```

```python
'''Modificando los fondos de los gr√°ficos, que tambi√©n se puede'''
fig, ax = plt.subplots(facecolor='xkcd:hot pink') #the figure facecolor will set the color of the outside margins 
ax.set_facecolor('deepskyblue') #the axis facecolor will set the color within the axis bounds

#axis line color
for spine in ['top','bottom','left','right']:
    ax.spines[spine].set_linewidth(3) #making them thicker to see them better
    ax.spines[spine].set_color('xkcd:golden yellow')
    
#ticks and their labels
ax.tick_params(length=6, width=4,color='xkcd:golden yellow', labelcolor='white',labelsize=10)
```

```python
# podemos usar rcParams para modificar los fondos de los gr√°ficos
import matplotlib as mpl
from cycler import cycler

mpl.rcParams['figure.facecolor'] = 'w'
mpl.rcParams['figure.edgecolor'] = '#cc6510'
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['axes.facecolor'] = 'y'
mpl.rcParams['axes.edgecolor'] = 'b'
mpl.rcParams['xtick.color'] = 'r'
mpl.rcParams['ytick.color'] = 'g'
mpl.rcParams['lines.color'] = 'white'
mpl.rcParams['axes.prop_cycle'] = cycler(linestyle=['-','--','-.',':'])
mpl.rcParams['lines.linewidth'] = 2

A = 1.
for i in range(3):
    plt.plot(x,A*y)
    A *= 0.6
plt.show()
```

```python
# una vez acabamos volvemos a poner los rcParams a default values
mpl.rcParams.update(mpl.rcParamsDefault)

plt.plot(x,y)
plt.show()
```
