# Matplotlib estilos en las líneas
Estos son los estilos  de líneas más utilizados aunque existen algunos más
+ '-' -> solid
+ '-.' -> dashdot
+ ':' -> dotted (punteada)
+ '--' -> dashed

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_eje = ['enero','febrero','marzo','abril','mayo','junio']
y_2020 = [123, 134, 156, 166, 177, 155]
y_2021 = [n*np.random.normal(1.06,0.1) for n in y_2020]
y_2022 = [n*np.random.normal(1.06, 0.1) for n in y_2021]
y_2023 = [n*np.random.normal(1.06, 0.1) for n in y_2022]
y_values = [y_2020, y_2021, y_2022, y_2023]
y_labels = [2020, 2021, 2022, 2023]
lines_styles = ['-','-.',':','--']
for y , j, k in zip(y_values, y_labels, lines_styles):
    plt.plot(x_eje, y,label = f'ventas {j}', linestyle=k)

plt.grid(linestyle='-.', linewidth=0.5, color='.25', zorder=-10)
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.title('Ventas por año')
plt.legend()
plt.show()
```
