# Matplotlib 
## Gráficos de líneas
Son los gráficos más básicos que podemos hacer. Consisten en representar puntos en coordenadas cartesianas (x,y) y unirlos mediente una línea.

En genral los gráficos de líneas permiten visualizar los cambios a lo largo de un rango continuo de una variable respecto a otra, como el tiempo o la distancia. La visualización del cambio con un gráfico de líneas permite ver de una sola vez la tendencia general y comparar simultáneamente varias tendencias. Nos sirven también para identificar facilmete máximos y mínimos en el intervalo estudiado.

Por ejemplo podemos definir en un rango( coordenadas x) varias funciones en el mismo gráfico(y1 = sen(x), y2 = cos(x), ...)

Para realizar la gráfica necesitaremos 2 arrays numéricos( np.array, list, tuple, ...). o más Cada uno contendrá: 
+ Valores en el eje X
+ Valores en el eje y
+ Puede haber más de una línea entonces tendremos valores en el eje y2
+ Valores en el eje y3 ...
Estos gráficos admiten además una serie de valores opcionales
+ Título
+ etiquetas ejex
+ etiquetas ejey
+ color de los puntos
+ color de las lineas
+ grosor de los puntos
+ tipos de líneas
+ incluso se puede cambiar los colores de fondo del gráfico

```python
# Importando librerías

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gráficos de líneas básicos
# Coordenadas ejes x e y
x_vals = np.arange(0, 10, 0.1)
y1_vals = np.sin(x_vals)
y2_vals = np.cos(x_vals)
```

```python
# Básico
plt.plot(x_vals, y1_vals)
plt.plot(x_vals, y2_vals)
```

```python
# Con opciones adicionales
plt.plot(x_vals, y1_vals, 'r--', lw=4, label = r'$y = \sin(x)$')
plt.plot(x_vals, y2_vals, 'b.-',label = r'$y = \sin(x)$')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.title('Funciones trigonométricas')
plt.legend()
```

```python
x_eje = ['enero','febrero','marzo','abril','mayo','junio']
y_2020 = [123, 134, 156, 166, 177, 155]
y_2021 = [n*np.random.normal(1.06,0.1) for n in y_2020]
y_2022 = [n*np.random.normal(1.06, 0.1) for n in y_2021]
y_2023 = [n*np.random.normal(1.06, 0.1) for n in y_2022]
```

```python

plt.plot(x_eje, y_2020, 'r--', lw=1, label = 'ventas 2020')
plt.plot(x_eje, y_2021, 'b--',label = 'ventas 2021')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.title('Funciones trigonométricas')
plt.legend()
```

```python
y_values = [y_2020, y_2021, y_2022, y_2023]
y_labels = [2020, 2021, 2022, 2023]
for y , j in zip(y_values, y_labels):
    plt.plot(x_eje, y,label = f'ventas {j}')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.title('Ventas por año')
plt.legend()
plt.show()
```

```python
# Observemos que podemos usar diferentes rangos tanto para x como para y
x_lista01=range(0,27)
x_lista02=range(0,20)
x_lista03=range(0,10)

y_lista01=[x**3+2 for x in x_lista01]
y_lista02=[2*x**2 + 6*x for x in x_lista02]
y_lista03=[200*x for x in x_lista03]

leyenda=[r'$x**3+2$',r'$2*x**2+ 6*x$', r'$200*x$']
plt.plot(x_lista01, y_lista01, '--', x_lista02, y_lista02, 'r--', x_lista03, y_lista03, 'g--', linewidth=2 )
plt.title('Varias funciones')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend(labels=leyenda)
plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')
plt.show()

```

## Usando datos de pandas
Matplotlib no solo acepta datos en formato numpy, list, tuple, o dictionary también acepta datos en formato pandas series.

Veamos un ejemplo:


```python
import pandas as pd
df = pd.read_csv('./data/covid_cv.csv')

# Creamos un gráfico de líneas
plt.plot(df.Mes,df.Fallecidos, marker='o', linestyle='dashed', linewidth=1.5, markersize=5, label='Fallecidos')
plt.plot(df.Mes,df.UCI, marker='v', linestyle='dashed', linewidth=1.5, markersize=5, label='UCI')
plt.title('Casos Covid-19 mensuales PV 2021')
plt.legend()
plt.xlabel('Mes')
plt.ylabel('Casos')
plt.show()
```

# Scatter plots
También son gráficos de los más básicos que podemos hacer. Consisten en representar puntos en coordenadas cartesianas (x,y). Concretamente es un gráfico que se usa para representar los valores de una variable sobre otra.

Nos sirven para tratar de encontrar una relación entre dos variables, por ejemplo si la relación es una relación lineal, cuadratica etc, o ayudarnos a identificar patrones o clusters en los datos.

```python

x_vals = np.arange(0, 10, 0.1)
y_vals = np.random.normal(1.06,0.1, 100)
y_vals_unif = np.random.uniform(2, 4, 100)
plt.scatter(x_vals, y_vals)
plt.xlabel(r'$x$')
plt.ylabel(r'$N(1, 0.1)$')
```

```python

from matplotlib import cm

plt.scatter(x_vals, y_vals_unif)
plt.xlabel(r'$x$')
plt.ylabel(r'$U(2,4)$')
```
