Curso básico de matplotlib.

+ 2. Configurar DPI y tamaño de gráficas

```python
# Importando librerías

import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# Configurando densidad de puntos(DPI) y figure size
plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 150
```

## Gráficos de líneas
Para crear gáficos de líneas, basicamete se usan puntos en coordenadas cartesianas (x,y). Por lo que necesitaremos un array con las coordenadasx y otro con las coordenadas y. Pueden ser arrays numpy, listas, tuplas ...

```python
# Gráficos de líneas básicos
# Coordenadas ejes x e y
x_vals = np.arange(0, 10, 0.1)
y_vals = np.sin(x_vals)

# plot básico
plt.plot(x_vals, y_vals)
```

## Opciones Adicionales
+ Etiquetas en los ejes `xlabel` e `ylabel`
+ Título del gráfico `title`
+ Etiquetas o labels
+ Color de las líneas
+ tipos de líneas

```python
# Veamoslo con un plot
y_vals_2 = np.cos(x_vals)
plt.plot(x_vals, y_vals, label = r'$y = \sin(x)$', color='r' )
plt.plot(x_vals, y_vals_2, label = r'$y = \sin(x)$', marker='.')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# color y marker se pueden combinar y abreviar
# Veamoslo con un plot
y_vals_2 = np.cos(x_vals)
plt.plot(x_vals, y_vals, 'r--',  label = r'$y = \sin(x)$')
plt.plot(x_vals, y_vals_2, 'b.-',label = r'$y = \sin(x)$')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# podemos modificar el grueso de línea
y_vals_2 = np.cos(x_vals)
plt.plot(x_vals, y_vals, 'r--', lw=4, label = r'$y = \sin(x)$')
plt.plot(x_vals, y_vals_2, 'b.-',label = r'$y = \sin(x)$')
plt.xlabel("$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# Veamos algunos estílos de líneas con diferentes markers
idx = 0
for marker in ('-', '--', ':', '-.', 'o', '.'):
    plt.plot(x_vals, np.sin(x_vals)+idx, ''.join(('b',marker)))
    idx += 1
```

```python
# Notar que el orden importa
shuffled_xvals = np.random.permutation(x_vals)
plt.plot(shuffled_xvals, np.sin(shuffled_xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Usando colormaps
plt.scatter(x_vals, np.sin(x_vals), c=np.cos(x_vals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```
