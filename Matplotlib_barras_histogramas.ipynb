# Gráficos de barras e histogramas 
## Gráfico de barras
Los gráficos de barras resumen y comparan los datos categóricos mediante longitudes de barras proporcionales a los valores que representas. Los gráficos de barras se componen de un eje x y un eje y.

Características:
+ Se utilizan para comparar dos o más valores.
+ Pueden ser verticales u horizontales
+ Permiten comparar conjuntos de datos por categorías

## Histograma
Un histograma es la representación gráfica en forma de barras, que simboliza la distribución de un conjunto de datos. Sirven para obtener una "primera vista" general, o panorama, de la distribución de una serie de datos (población, o muestra), respecto a una característica, cuantitativa y continua.

Características:
+ Utiliza barras verticales, cada barra o rectángulo indica la frecuencia de cada valor
+ No hay que confundir con las gráficas de barras
+ en matplotlib simplemete se le pasan los datos y ya realiza los cálculor para ralizar la gráfica

Los histogramas se usan en datos continuos, mientras que los diagramas de barras se usan en datos categóricos o nominales.


```python
# iportamos librerías necesarias
# numpy for math/array operations
# matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Creamos los datos
nums = np.random.uniform(size=10)
nums
```

```python
# Barplots

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

plt.bar(fruits, counts, label=bar_labels, color=bar_colors)

plt.ylabel('fruit supply')
plt.title('Fruit supply by kind and color')
plt.legend(title='Fruit color')

plt.show()
```

```python
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


#fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = plt.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    plt.bar_label(p, label_type='center')

plt.title('Number of penguins by sex')
plt.legend()

plt.show()
```

```python

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count
ax.bar_label(p, label_type='center')
ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```
